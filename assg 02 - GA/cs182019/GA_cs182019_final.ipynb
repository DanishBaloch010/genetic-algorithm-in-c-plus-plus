# 1) Function for finding best solution
# 2) Function for updating relative fitness

```python
# Finding best Fitness , Calc all relative percentages , Calc child fitness
# function to find best solution
def find_max_fit(list1):
    maximum = list1[0][1]
    for i in range(len(list1)):
        if list1[i][1] > maximum:
            maximum = list1[i][1]
            max_index = i
            
    return maximum , max_index

# function to calc all the fit_perc
def calc_all_perc(list1 , totalfit):
    for i in range(len(list1)):
        fit1 = list1[i][1]
        percentage = round((fit1/totalfit)*100 ,1)
        list1[i][2] = percentage
    return list1 , totalfit

```

# 1) Function for calculating fitness of string
# 2) Function for calculating fitness of initial population

```python
# Fitness Function
# function for calculating fitness for given string
def fitness(string):
    final_ans = 0
    position = 1
    temp_arr = list(string)
    for i in range(len(temp_arr)):
        str1 = temp_arr[i]
        final = str1 + str(position)
#         searching the orginal 2d array
        for x in range(len(array)):
            for y in range(len(array[x])):
                arr = array[x][y]
                temp = arr.split("-")
                if temp[1] == final:
                    units_sold = temp[0]
                    final_ans = final_ans + int(units_sold)
                    position = position + 1

    return final_ans

# Inital Population Fitness Calculation
# function for calculating intial population fitness
def calc_intial_pop_fitness(list_of_sol , total , list_final_fit):
    list_fitness = []
#     calc fitness and storing string , fitness
    for i in range(len(list_of_sol)):
        fit = 0
        fit = fitness(list_of_sol[i])
        list_fitness.append([list_of_sol[i],fit])
        total = total + fit

    for i in range(len(list_fitness)):
        string = list_fitness[i][0]
        fit = list_fitness[i][1]
        perc = round((fit/total)*100 ,1)
        list_final_fit.append([string, fit, perc])
        
    return list_final_fit , total
```

## --> also checks if mutated string is feasible or not

```python

# Mutation Function
def mutation(string, total, list1):
    while True:
        string_arr = list(string[0])
        mutate_what_counter = 0
        mutate_with_counter = 0
        while(mutate_what_counter == 0):
            mutate_what = str(input("Mutate what from string A,B,C,D,E ?"))
            mutate_what = mutate_what.upper()
            for i in range(len(string_arr)):
                if string_arr[i] == mutate_what:
                    mutate_what_counter = mutate_what_counter + 1
                    print("Mutate what is present in string.")
                    break
            if(mutate_what_counter != 1):
                print("ERROR ! Mutate what is not present in string, try again")
    
        possible_selections =['A','B','C','D','E']
        while(mutate_with_counter != 1):
            mutate_with = str(input("Mutate with A,B,C,D,E ?"))
            mutate_with = mutate_with.upper()
            for i in range(len(possible_selections)):
                if possible_selections[i] == mutate_with:
                    mutate_with_counter = mutate_with_counter + 1
                    print("Mutate with is present in  SELECTIONS.")
                    break
            if(mutate_with_counter != 1):
                print("ERROR ! Mutate with is not present in  SELECTIONS, try again")

        temp_arr = list(string[0])
        for i in range(len(temp_arr)):
            temp_str = temp_arr[i]
            if mutate_what == temp_str:
                temp_arr[i] = mutate_with
                break
    
        mutated_string = ""
        for i in range(len(temp_arr)):
            mutated_string = mutated_string + temp_arr[i]
        
        mutated_string = mutated_string.upper()
        
        valid_mutate = True
        for i in range(len(mutated_string)):
            first_char =  mutated_string[i]
            for j in range(len(mutated_string)):
                if i != j :
                    if mutated_string[j] == first_char:
                        valid_mutate = False
                        
        if valid_mutate == True:
            print("feasible mutate --> " , mutated_string)
        elif valid_mutate == False:
            print("infeasible mutate --> " , mutated_string)
            
        if(valid_mutate == True):
            break
        
    fit_val = fitness(mutated_string)
    total = total + fit_val
    perc = round((fit_val/total)*100,1)
    list1.append([mutated_string,fit_val,perc])
    list1 , total = calc_all_perc(list1 , total)
    
    return mutated_string , total , list1

# fucntion to select one string for mutation
def mutation_select(tot , temp_list):
    mutate_list = []
    counter = 0
    while(counter != 1):
        check = False
        sol_string = str(input("Enter a string from mating pool for mutation: "))
        sol_string = sol_string.upper()
        for i in range(len(list_final)):
            if list_final[i][0] == sol_string:
                check = True
                temp_str = list_final[i][0]
                temp_fit_val = list_final[i][1]
                temp_fit_perc = list_final[i][2]
                mutate_list.append([temp_str, temp_fit_val, temp_fit_perc])
                counter = counter + 1
                break
                
        if check == False:
            print("string not available in mating pool for mutation")
    
    str1 = mutate_list[0]
    mutated , tot , temp_list = mutation(str1 , tot , temp_list)
    
    return mutated , tot , temp_list
```
